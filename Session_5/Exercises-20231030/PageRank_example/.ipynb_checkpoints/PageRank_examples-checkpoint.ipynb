# PageRank algorithm

## Basic PageRank

<img src="base_example.png" alt="Drawing" style="width: 500px;" align="left"/>

```python
M = np.array([[0,    0.5,  1,  0  ],
              [1/3,  0,    0,  0.5],
              [1/3,  0,    0,  0.5],
              [1/3,  0.5,  0,  0  ]])

v = np.array([0.25, 0.25, 0.25, 0.25])
```

```python
n_steps = 1000000

for i in range(n_steps):
    v_old = v.copy()
    v = M @ v
    
    if np.isclose(v, v_old).all():
        print(f'PageRank converged at iteration {i}')
        break
```

```python
v
```

## Taxation / teleportation

```python
M = np.array([[0,    0.5,  0,  0  ],
              [1/3,  0,    0,  0.5],
              [1/3,  0,    1,  0.5],
              [1/3,  0.5,  0,  0  ]])

v = np.array([0.25, 0.25, 0.25, 0.25])
```

```python
n_steps = 100
beta = 0.8

n = v.shape[0]
e = np.ones(n)

for i in range(n_steps):
    v_old = v.copy()
    v = beta * (M @ v) + (1 - beta) * (1/n) * e

    if np.isclose(v_old, v).all():
        print(f'PageRank coverged at iteration {i}')
        break
```

```python
v
```

## Exercise 2

```python
M = np.array([[1/3,  0.5,  0  ],
              [1/3,  0,    0.5],
              [1/3,  0.5,  0.5]])

v = np.array([1/3, 1/3, 1/3])
```

```python
n_steps = 1000000
beta = 0.8

n = v.shape[0]  # number of pages in the graph
e = np.ones(n)

for i in range(n_steps):
    v_old = v.copy()
    v = beta * (M @ v) + (1 - beta) * (1/n) * e
    
    if np.isclose(v_old, v).all():
        print(f'PageRank coverged at iteration {i}')
        break
```

```python
v
```

```python
print(f'r1 = {21/81}, r2 = {25/81}, r3 = {35/81}')
```

# PageRank with NetworkX

```python
import itertools
import random

import networkx as nx
from matplotlib import pyplot as plt
```

## Synthetic data

```python
nodes = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

```python
G = nx.Graph()
G.add_nodes_from(nodes)
```

```python
for (node1, node2) in itertools.combinations(nodes, r=2):
    if random.uniform(0,1) < 0.5:
        G.add_edge(node1, node2)
```

```python
# Draw generated graph
nx.draw_networkx(G, pos=nx.circular_layout(G), with_labels=True)
```

```python
nx.pagerank(G)
```

## Real data
